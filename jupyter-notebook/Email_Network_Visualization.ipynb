# Basic Visualization of the Email Network

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import networkx as nx
from networkx.algorithms import bipartite
```

```python
#Load Edge List
file = '../data/initial/email-edges.txt'
```

```python
G = nx.read_edgelist(file)
```

```python
nx.draw(G, node_size=10)
plt.show()
```

```python
nx.draw_spring(G, node_size=5, node_color='navy', width=0.8, edge_color='silver')
file='../image/Network_General.png'
plt.savefig(file)
plt.show()
```
